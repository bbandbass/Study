# 뉴스그룹 데이터에 대한 이해

```python
import numpy as np
import pandas as pd
from sklearn.datasets import fetch_20newsgroups
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
```

sklearn에서 제공하는 twenty newsgroups라 불리는 20개 다른 주제에 대한 뉴스그룹 데이터 제공. LSA는 문서 검색에 조금 더 특화된 알고리즘이지만 토픽 모델링 분야의 시초가 된 알고리즘이기에, LSA를 사용해서 문서의 수를 토픽의 수로 압축한 뒤, 각 토픽당 가장 중요한 5개 단어를 출력하여 토픽 모델링 수행

```python
dataset = fetch_20newsgroups(shuffle=True, random_state=1, remove=('headers', 'footers', 'quotes'))
documents = dataset.data
print('샘플의 수 :',len(documents))
```

본래 이 뉴스그룹 데이터가 가지고 있던 20개의 카테고리

# 텍스트 전처리

```python
news_df = pd.DataFrame({'document':documents})
# 정규표현식(알파벳이 아닌 모든 문자)을 이용하여 특수 문자 제거
news_df['clean_doc'] = news_df['document'].str.replace("[^a-zA-Z]", " ")

# 길이가 3이하인 단어는 제거 (길이가 짧은 단어 제거)
news_df['clean_doc'] = news_df['clean_doc'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>3]))

# 전체 단어에 대한 소문자 변환(단어의 개수를 줄여줌)
news_df['clean_doc'] = news_df['clean_doc'].apply(lambda x: x.lower())
```

```python
news_df['clean_doc'][1]
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')
tokenized_doc = news_df['clean_doc'].apply(lambda x: x.split()) # 공백을 기준으로 토큰화
tokenized_doc = tokenized_doc.apply(lambda x: [item for item in x if item not in stop_words])
# 불용어가 아닌 단어들만 남기기
```

```python
print(tokenized_doc[1])
```

## TF-IDF 행렬 만들기

불용어 제거를 위해 토큰화를 수행했지만, TF-IDF는 토큰화 되어있지 않는 텍스트 데이터를 입력으로 사용.
따라서 토큰화 작업을 역으로 취소하는 역토큰화 수행

```python
# 역토큰화 (토큰화 작업을 역으로 되돌림)
detokenized_doc = []
for i in range(len(news_df)):
    t = ' '.join(tokenized_doc[i]) # 공백을 term들 사이에 두며 join
    detokenized_doc.append(t)

news_df['clean_doc'] = detokenized_doc
```

```python
news_df['clean_doc'][1]
```

```python
vectorizer = TfidfVectorizer(stop_words='english', max_features= 1000, # 상위 1,000개의 단어를 보존 
max_df = 0.5, smooth_idf=True)

# max_df = 0.5 단어 t가 등장한 문서가 전체 문서의 절반 이상이면 해당 단어 무시
# idf 구할 때 분모에 1 더해서 0으로 나뉘는 것을 방지해줌

X = vectorizer.fit_transform(news_df['clean_doc'])

# TF-IDF 행렬의 크기 확인
print('TF-IDF 행렬의 크기 :',X.shape)
```

# Topic Modeling

기존의 뉴스그룹 카테고리가 20개였기 때문에 20개의 topic이 존재할 것이라 가정하고 토픽 모델링

```python
svd_model = TruncatedSVD(n_components=20, algorithm='randomized', n_iter=100, random_state=122)
svd_model.fit(X)
len(svd_model.components_)
```

```python
np.shape(svd_model.components_)
```

토픽의 수 x 단어의 수를 가지는 기존에 설명했던 잠재의미를 통해 단어벡터를 설명하는 행렬

여기서는 기존에 설명했던 것과 달리 U, V를 바꾸어 설명하고 있기에 V^T에 해당하는 행렬

```python
terms = vectorizer.get_feature_names() # 단어 집합. 1,000개의 단어가 저장됨.

def get_topics(components, feature_names, n=5):
    for idx, topic in enumerate(components):
        print("Topic %d:" % (idx+1), [(feature_names[i], topic[i].round(5)) for i in topic.argsort()[:-n - 1:-1]])
get_topics(svd_model.components_,terms)
```
